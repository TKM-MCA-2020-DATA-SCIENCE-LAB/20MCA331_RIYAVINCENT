```python
from sklearn.model_selection  import train_test_split
```

```python
x=new_data#=dt.Class
y=dt['Class']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size= 0.3,random_state=42)
```

```python
 from sklearn.neighbors  import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)# k=1
```

```python
knn.fit(x_train,y_train)    #training the dataset using x_train and y_train
```

```python
#predicting
pred=knn.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix #evaluation matrix-
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
error_rate=[]
for i in range(1,100):
  knn=KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train,y_train)
  pred_i=knn.predict(x_test)
  error_rate.append(np.mean(pred_i !=y_test))
```

```python
plt.figure(figsize=(10,6)) 
plt.plot(range(1,100),error_rate,color='black',linestyle='dashed',marker='s',markerfacecolor='yellow',markersize=10)
plt.title('error rate vs values')
plt.xlabel('k')
plt.ylabel('error rate')
```
